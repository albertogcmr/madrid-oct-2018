### Mejora de los datos

Una vez eliminadas las columnas cuyos datos son inválidos y los registros que contienen valores también inválidos para nosotros o que no aporten información pasamos a mejorar la calidad de los datos

Con el ejemplo de diferentes actividades

```python
activities = data_na_clean.Activity.value_counts()
activities.head(20)
```

Se aprecia que **data_na_clean** todavía contiene registros que pueden considerarse duplicados como: Scuba diving <-> Free diving

```python
# NOTA: este proceso se puede realizar con pandas de una forma más fácil. 
```

```python
# Ejemplo de diccionario para traducir actividades

# NOTA: Visitar
# http://pandas.pydata.org/pandas-docs/stable/indexing.html#indexing-view-versus-copy
CONTAINS_ALIAS_ACTIVITY = {
    'Surfing': ['Surf', 'surf', 'board', 'Board'], 
    'Swimming': ['Swim', 'swim', 'snor', 'Snor', 'float', 'Float', 'Bath', 'bath'], 
    'Fishing': ['Fish', 'fish'], 
    'Diving': ['Divi', 'divi']
}
for key, value in CONTAINS_ALIAS_ACTIVITY.items():
    for act in value: 
        data_na_clean.loc[data_na_clean['Activity'].str.contains(act, regex=False), 'Activity'] = key
```

```python
activities = data_na_clean.Activity.value_counts()
activities.head(20)
```

Se puede aprceiar que se han ido agrupando los Surfing. A medida que se incremente el diccionario CONTAINS_ALIAS_ACTIVITY irá mejorando la solución


Con el ejemplo de diferentes tipos de ataque podemos hacer algo similar a lo realizado con actividades. 

```python
types = data_na_clean.Type.value_counts()
types
```

```python
CONTAINS_ALIAS_TYPE = {
    'Boating': ['Boat', 'boat']
}

# mejorable 
for key, value in CONTAINS_ALIAS_TYPE.items():
    for ty in value: 
        data_na_clean.loc[data_na_clean['Type'].str.contains(ty, regex=False), 'Type'] = key
```

```python
types = data_na_clean.Type.value_counts()
types
```

Podemos apreciar que Boating ahora agrupa Boat y Boating

#### 2. Agrupamos los paises por la cantidad ataques sufridos en el histórico

```python
df_paises_bin = data_na_clean['Country'].value_counts()
df_paises_bin.head(20)
```

```python
# NOTA: leer documentación es importante para entender los errores

# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.qcut.html
"""
duplicates : {default ‘raise’, ‘drop’}, optional
If bin edges are not unique, raise ValueError or drop non-uniques.
New in version 0.20.0.
"""

mpg_labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']
bins = pd.qcut(df_paises_bin,6, labels=mpg_labels, duplicates='drop')
bins
```

Hemos seleccionado qcut porque nuestra distribución tiene muchos valores atípicos (outliers): USA y Australia muchos ataque pero la gran mayoría del resto muy pocos
