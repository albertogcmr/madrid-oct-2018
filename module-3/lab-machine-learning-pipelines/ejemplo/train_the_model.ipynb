```python
import pandas.io
import pandas as pd

def readCsvToDataFrame(path, row_with_column_names):
    theDataFrame = pandas.read_csv(path, names=row_with_column_names)
    return theDataFrame.astype('float64')

colnames =[
    "1_class", 
    "2_lymphatics" ,
    "3_block_of_affere",
    "4_bl_of_lymph_c",
    "5_bl_of lymph_s",
    "6_by_pass",
    "7_extravasates",
    "8_regeneration_of", 
    "9_early_uptake_in",
    "10_lym_nodes_dimin",
    "11_lym_nodes_enlar",
    "12_changes_in_lym",
    "13_defect_in node", 
    "14_changes_in_node", 
    "15_changes_in_stru",
    "16_special_forms",
    "17_dislocation",
    "18_exclusion",
    "19_no_of_nodes_in",
]
df = readCsvToDataFrame('./lymphography.data',colnames)
df.head(3)
```

```python
def show_df_info(dataframe):
    # get the data type
    print(type(dataframe))
    print("amount of entries is %s" % dataframe.size)
    print("dimensions= %i" % dataframe.ndim)
    print("shape is ", end="")
    print(dataframe.shape)
    print("axes: ", end="")
    print(dataframe.axes)
    print("data types of columns:")
    print(dataframe.dtypes)
    print("features: %s" % dataframe.columns)

show_df_info(df)
```

```python
def sliceDataFrame(df):
    # remove 2 instances with class label "normal find"
    df= df.drop(df[df['1_class']==1].index)
    # remove 4 instances with class label "fibrosis"
    df = df.drop(df[df['1_class'] == 4].index)
    # iloc arguments: range of rows, range of columns
    # class 'pandas.core.series.Series'
    return df.iloc[:, 1:], df.iloc[:, 0]

X_data, y = sliceDataFrame(df)
display(X_data.head(3))
display(y.head(3))
```

```python
# CREATE THE PIPELINE
from sklearn import tree, pipeline, preprocessing
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

cart_model = tree.DecisionTreeClassifier()

pipe = pipeline.Pipeline(steps=[
    ("feature_selection", SelectKBest(chi2, k=10)),
    ("scale", preprocessing.StandardScaler()),  
    ("CART", cart_model)
])


pipe.fit(X_data, y=y)
```

```python
from sklearn.model_selection import cross_val_score

# return value is array of scores
scores = cross_val_score(pipe, X_data, y, cv=50)
display(scores)
# use as quality metric the average CV score
meanCvAccuracy = scores.mean()

print("Mean CV accuracy= %f" % meanCvAccuracy)
```
